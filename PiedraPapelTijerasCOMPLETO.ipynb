```python
import tkinter as tk
from tkinter import messagebox
import random

class PiedraPapelTijera:
    def __init__(self, root):
        self.root = root
        self.root.title("Piedra, Papel, Tijera - Extendido")

        # Opciones basadas en la imagen proporcionada (sin Pirata)
        self.opciones = [
            ("Piedra", "ü™®"),
            ("Papel", "üìÑ"),
            ("Tijera", "‚úÇÔ∏è"),
            ("Lagarto", "ü¶é"),
            ("Spock", "üññ"),
            ("Heavy", "üí™"),
            ("Papa", "ü•î")
        ]

        self.ganados = 0
        self.perdidos = 0
        self.empates = 0
        self.crear_interfaz()

    def crear_interfaz(self):
        # Colores de fondo
        bg_color = "#e0e7ff"
        frame_color = "#c7d2fe"
        self.root.configure(bg=bg_color)

        # T√≠tulo
        titulo = tk.Label(self.root, text="PIEDRA PAPEL TIJERA EXTENDIDO",
                         font=("Arial", 16, "bold"), bg=bg_color)
        titulo.pack(pady=10)

        # Resultado
        self.resultado = tk.Label(self.root, text="Elige una opci√≥n para comenzar",
                                 font=("Arial", 12), bg=bg_color, wraplength=400)
        self.resultado.pack(pady=10)

        # Botones de opciones
        frame_botones = tk.Frame(self.root, bg=frame_color)
        frame_botones.pack(pady=10)

        self.botones = []
        for i, (nombre, emoji) in enumerate(self.opciones):
            btn = tk.Button(frame_botones, text=f"{emoji}\n{nombre}",
                          width=10, height=3, font=("Arial", 9),
                          command=lambda n=nombre: self.jugar(n),
                          bg="#dbeafe", relief="raised", bd=2)
            btn.grid(row=i//4, column=i%4, padx=5, pady=5)
            self.botones.append(btn)

        # Marcador
        self.marcador = tk.Label(self.root,
                               text="Victorias: 0 | Derrotas: 0 | Empates: 0",
                               font=("Arial", 11, "bold"), bg=bg_color)
        self.marcador.pack(pady=10)

        # Bot√≥n de reinicio
        btn_reiniciar = tk.Button(self.root, text="Reiniciar Juego",
                                font=("Arial", 10), command=self.reiniciar,
                                bg="#fef3c7", relief="raised", bd=2)
        btn_reiniciar.pack(pady=5)

    def jugar(self, eleccion_usuario):
        eleccion_bot_nombre, eleccion_bot_emoji = random.choice(self.opciones)
        eleccion_bot = eleccion_bot_nombre

        # Encontrar emoji del usuario
        usuario_emoji = next(emoji for nombre, emoji in self.opciones if nombre == eleccion_usuario)

        resultado = self.determinar_ganador(eleccion_usuario, eleccion_bot)

        # Actualizar contadores
        if resultado == "¬°Ganaste!":
            self.ganados += 1
            color = "green"
        elif resultado == "Perdiste":
            self.perdidos += 1
            color = "red"
        else:
            self.empates += 1
            color = "blue"

        # Actualizar interfaz
        self.resultado.config(
            text=f"{resultado}\n\nT√∫: {usuario_emoji} {eleccion_usuario}\nBot: {eleccion_bot_emoji} {eleccion_bot}",
            fg=color
        )
        self.marcador.config(
            text=f"Victorias: {self.ganados} | Derrotas: {self.perdidos} | Empates: {self.empates}"
        )

    def determinar_ganador(self, usuario, bot):
        if usuario == bot:
            return "Empate"

        # Reglas actualizadas sin Pirata
        reglas = {
            "Piedra": ["Tijera", "Lagarto", "Heavy", "Papa"],
            "Papel": ["Piedra", "Spock", "Heavy"],
            "Tijera": ["Papel", "Lagarto", "Heavy", "Papa"],
            "Lagarto": ["Spock", "Papel", "Heavy"],
            "Spock": ["Tijera", "Piedra", "Heavy", "Papa"],
            "Heavy": ["Spock", "Lagarto", "Tijera"],
            "Papa": ["Papel", "Lagarto", "Heavy"]
        }

        return "¬°Ganaste!" if bot in reglas[usuario] else "Perdiste"

    def reiniciar(self):
        self.ganados = 0
        self.perdidos = 0
        self.empates = 0
        self.resultado.config(text="Elige una opci√≥n para comenzar", fg="black")
        self.marcador.config(text="Victorias: 0 | Derrotas: 0 | Empates: 0")

if __name__ == "__main__":
    root = tk.Tk()
    app = PiedraPapelTijera(root)
    root.mainloop()
```
